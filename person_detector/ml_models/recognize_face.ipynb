```python
from PIL import Image

from keras_facenet import FaceNet
import numpy as np
from numpy import asarray
from numpy import expand_dims

import datetime
import time
import pickle
import cv2
```

```python
HaarCascade = cv2.CascadeClassifier(cv2.samples.findFile(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'))
MyFaceNet = FaceNet()
```

```python
cap = cv2.VideoCapture(0)
# address = 'http://192.168.242.101:8080/video'
# cap.open(address)

def make_1080p():
    cap.set(3, 1920)
    cap.set(4, 1080)

def make_720p():
    cap.set(3, 1280)
    cap.set(4, 720)

# def make_480p():
#     cap.set(3, 640)
#     cap.set(4, 480)

# make_720p()
last_print_time = time.time()
while(1):
    myfile = open("data.pkl", "rb")
    database = pickle.load(myfile)
    myfile.close()

    _, gbr1 = cap.read()
    
    wajah = HaarCascade.detectMultiScale(gbr1,1.1,4)
    
    if len(wajah)>0:
        x1, y1, width, height = wajah[0]        
    else:
        x1, y1, width, height = 1, 1, 10, 10
    
    x1, y1 = abs(x1), abs(y1)
    x2, y2 = x1 + width, y1 + height
    
    
    gbr = cv2.cvtColor(gbr1, cv2.COLOR_BGR2RGB)
    gbr = Image.fromarray(gbr)                  # konversi dari OpenCV ke PIL
    gbr_array = asarray(gbr)
    
    face = gbr_array[y1:y2, x1:x2]                        
    
    face = Image.fromarray(face)                       
    face = face.resize((160,160))
    face = asarray(face)
    
#     face = face.astype('float32')
#     mean, std = face.mean(), face.std()
#     face = (face - mean) / std
    
    face = expand_dims(face, axis=0)
#     signature = MyFaceNet.predict(face)
    signature = MyFaceNet.embeddings(face)

    
    min_dist=100
    identity=' '
    for key, value in database.items() :
        dist = np.linalg.norm(value-signature)
        if dist < min_dist:
            min_dist = dist
            identity = key
            
    if identity and time.time() - last_print_time >= 2:
        current_time = datetime.datetime.now().strftime('%H:%M:%S')
        message = f"Nama wajah terdeteksi: {identity}; Waktu: {current_time}"
        print(message)
        last_print_time = time.time()  # Memperbarui waktu terakhir pesan dicetak
        
    cv2.putText(gbr1,identity, (100,100),cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2, cv2.LINE_AA)
    # print("Nama Wajah Terdeteksi:", identity)
    cv2.rectangle(gbr1,(x1,y1),(x2,y2), (0,255,0), 2)
        
    cv2.imshow('face_recognition',gbr1)
    
    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break
        
cv2.destroyAllWindows()
cap.release()
```
